```python
test = []

for page in range(0, 62):
    url = 'https://www.korcham.net/nCham/Service/EconBrief/appl/ProspectBoardList.asp?board_type=1&pageno={}&daybt=OldNow&m_OldDate=20170901&m_NowDate=20210602'
    req = requests.get(url.format(str(page)))
    soup = BeautifulSoup(req.content.decode('cp949', 'replace'), 'html.parser')
    
    temp = soup.find_all('table')
    print(temp)
    
    p = parser.make2d(temp[0])
    table = pd.DataFrame(p[1:], columns= p[0])
    test.append(table)
```

```python
len(test)
```

```python
list_ko = []

for i in range(1, 62):
    df = pd.DataFrame(test[i])
    print(df)
    list_ko.append(df)
```

```python
df_result = pd.concat(list_ko)
df_result2 = df_result.sort_values(by=['날짜'], axis=0, ascending=True)


df_result2 .reset_index(drop = True, inplace =True)
df_result2
```

```python
df_result2.to_csv('../data/시장금리_일별.csv', encoding= 'UTF-8-sig')
```
